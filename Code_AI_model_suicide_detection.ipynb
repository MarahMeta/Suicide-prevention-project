```python

from transformers import pipeline

pipe = pipeline("text-classification", model="vibhorag101/roberta-base-suicide-prediction-phr")
```

Get a feeling for the model with an example:

```python
pipe("I cant take this anymore")
```

Deploy model

```python
def deploy_model(input):
  pipe(input)
  return pipe (input)



def format_output(output):
    # Assuming output is a list with one dictionary inside
    if output[0]['label'] == 'non-suicide':
        background_color = 'green'
    else:
        background_color = 'red'

    label = output[0]['label']
    return f'<div style="background-color: {background_color}; padding: 5px; display: inline-block;">{label}</div>'



demo = gr.Interface(fn=deploy_model, inputs="text", outputs="text", title= "AI model for suicide detection", description="Paste in text, e.g. a post or tweet and the model will evaluate whether this content may indicate suicidal tendencies.")
demo.launch(share=True, debug=True)
```
